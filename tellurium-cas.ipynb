# CombineShowCase archive with tellurium


## Download the latest archive
https://github.com/SemsProject/CombineArchiveShowCase

```python
from __future__ import print_function
import requests
import shutil

url = "http://scripts.sems.uni-rostock.de/getshowcase.php"
cas_path = "./results/cas.omex"
response = requests.get(url, stream=True)
with open(cas_path, 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)
del response

print("CombinArchiveShowcase downloaded: ", cas_path)
```

## Run archive with tellurium

```python
import tellurium
from tellurium.sedml import tesedml

# interactive plotly results
# tellurium.setDefaultPlottingEngine("plotly")
# tesedml.executeCombineArchive(cas_path)

# store the matplotlib figures
tellurium.setDefaultPlottingEngine("matplotlib")
tesedml.executeCombineArchive(cas_path, saveOutputs=True, outputDir="./results/output");
```
